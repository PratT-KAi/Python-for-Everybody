# Dictionaries

```python
di = {}
di['age'] = 21
di['course'] = 101
di
```

```python
names = ['john', 'gwen', "sofi", 'monty', 'john']
freq = {}
for i in names:
    freq[i] = freq.get(i,0) + 1
#     if i in freq: freq[i] +=1    
#     else: freq[i] = 1
freq
```

```python
freq.get('pratik',0)
for i in freq:
    print(i,freq[i])
```

```python
print(list(freq))
print(freq.keys())
print(freq.values())
print(freq.items())
```

```python
for i,j in freq.items():
    print(i,j)
```
